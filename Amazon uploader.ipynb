```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
import os

# Load the dataset
df = pd.read_csv('F:/MDTM46B/Project 4/Amazon Music Clustering/single_genre_artists.csv')
print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print("\nFirst 3 rows:")
print(df.head(3))

# Check basic info
print("\n" + "="*50)
print("DATASET INFORMATION")
print("="*50)
print(f"Number of songs: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
print(f"Columns: {list(df.columns)}")

# Check missing values
print("\n" + "="*50)
print("MISSING VALUES")
print("="*50)
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

# Check data types
print("\n" + "="*50)
print("DATA TYPES")
print("="*50)
print(df.dtypes)

# Drop unnecessary columns for clustering
columns_to_drop = ['id_songs', 'name_song', 'explicit', 'id_artists', 
                   'release_date', 'key', 'mode', 'time_signature', 
                   'followers', 'genres', 'name_artists', 'popularity_artists']

# Keep only numeric audio features
audio_features = ['danceability', 'energy', 'loudness', 'speechiness', 
                  'acousticness', 'instrumentalness', 'liveness', 'valence', 
                  'tempo', 'duration_ms', 'popularity_songs']

df_clean = df[audio_features].copy()
print(f"\nClean dataset shape: {df_clean.shape}")
print(f"Selected features: {list(df_clean.columns)}")

# Handle any missing values (fill with median)
df_clean = df_clean.fillna(df_clean.median())
print(f"\nMissing values after cleaning: {df_clean.isnull().sum().sum()}")

# Save clean dataset
os.makedirs('outputs', exist_ok=True)
df_clean.to_csv('outputs/clean_audio_features.csv', index=False)
print("\n✅ Clean dataset saved!")

# Quick visualization of feature distributions
plt.figure(figsize=(15, 10))
for i, col in enumerate(audio_features, 1):
    plt.subplot(3, 4, i)
    plt.hist(df_clean[col], bins=30, alpha=0.7, color='skyblue')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('outputs/feature_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n📊 Feature distributions saved as 'feature_distributions.png'")
```
