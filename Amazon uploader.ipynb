```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import joblib

# Load scaled data
df_scaled = pd.read_csv('outputs/scaled_audio_features.csv')
print(f"Loaded scaled data: {df_scaled.shape}")

# Method 1: Elbow Method to find optimal K
print("\n" + "="*50)
print("ELBOW METHOD - Finding Optimal Number of Clusters")
print("="*50)

inertia_values = []
k_range = range(2, 15)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal k')
plt.grid(True, alpha=0.3)
plt.axvline(x=6, color='red', linestyle='--', alpha=0.7, label='Suggested k=6')
plt.legend()
plt.tight_layout()
plt.savefig('outputs/elbow_method.png', dpi=300, bbox_inches='tight')
plt.show()

# Method 2: Silhouette Score
print("\n" + "="*50)
print("SILHOUETTE SCORE ANALYSIS")
print("="*50)

silhouette_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(df_scaled)
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot Silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, 'ro-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different k Values')
plt.grid(True, alpha=0.3)
plt.axvline(x=6, color='red', linestyle='--', alpha=0.7, label='Suggested k=6')
plt.legend()
plt.tight_layout()
plt.savefig('outputs/silhouette_scores.png', dpi=300, bbox_inches='tight')
plt.show()

# Choose optimal k (let's go with k=6 based on both methods)
optimal_k = 6
print(f"\n🎯 Optimal number of clusters selected: {optimal_k}")

# Apply K-Means with optimal k
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(df_scaled)

# Add cluster labels to original data
df_results = df_scaled.copy()
df_results['cluster'] = cluster_labels

print(f"\nCluster distribution:")
print(df_results['cluster'].value_counts().sort_index())

# Calculate silhouette score for final model
final_silhouette = silhouette_score(df_scaled, cluster_labels)
print(f"\nFinal Silhouette Score: {final_silhouette:.3f}")

# Save the clustering model
joblib.dump(kmeans_final, 'outputs/kmeans_model.pkl')
print("\n✅ K-Means model saved!")

# Save results with clusters
df_results.to_csv('outputs/songs_with_clusters.csv', index=False)
print("\n✅ Clustering results saved!")

# Quick cluster size visualization
plt.figure(figsize=(8, 6))
cluster_counts = df_results['cluster'].value_counts().sort_index()
plt.bar(cluster_counts.index, cluster_counts.values, color='skyblue', edgecolor='navy')
plt.xlabel('Cluster')
plt.ylabel('Number of Songs')
plt.title(f'Cluster Sizes (k={optimal_k})')
plt.xticks(range(optimal_k))
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/cluster_sizes.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n📊 All clustering files saved successfully!")
print(f"Total songs clustered: {len(df_results)}")
print(f"Number of clusters: {optimal_k}")
print(f"Average silhouette score: {final_silhouette:.3f}")
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Load data
df_results = pd.read_csv('outputs/songs_with_clusters.csv')
df_original = pd.read_csv('data/music_data.csv')

print(f"Loaded data with clusters: {df_results.shape}")
print(f"Original data: {df_original.shape}")

# Prepare data for visualization - merge with original for track names
df_viz = df_results.copy()
df_viz['track_name'] = df_original['name_song'].values[:len(df_viz)]
df_viz['artist_name'] = df_original['name_artists'].values[:len(df_viz)]

# 1. PCA for 2D visualization
print("\n" + "="*50)
print("2D CLUSTER VISUALIZATION USING PCA")
print("="*50)

pca = PCA(n_components=2, random_state=42)
pca_result = pca.fit_transform(df_results.drop('cluster', axis=1))

# Create PCA dataframe
df_pca = pd.DataFrame({
    'PC1': pca_result[:, 0],
    'PC2': pca_result[:, 1],
    'cluster': df_results['cluster']
})

# Plot PCA scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], 
                     c=df_pca['cluster'], cmap='tab10', alpha=0.7, s=30)
plt.xlabel(f'First Principal Component (PC1) - {pca.explained_variance_ratio_[0]:.1%} variance')
plt.ylabel(f'Second Principal Component (PC2) - {pca.explained_variance_ratio_[1]:.1%} variance')
plt.title('Song Clusters in 2D Space (PCA)')
plt.colorbar(scatter, label='Cluster')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/pca_clusters.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"PCA explained {sum(pca.explained_variance_ratio_):.1%} of variance")

# 2. Feature importance per cluster
print("\n" + "="*50)
print("FEATURE PROFILES BY CLUSTER")
print("="*50)

# Calculate mean features per cluster
cluster_profiles = df_results.groupby('cluster').mean()
print("\nAverage feature values by cluster:")
print(cluster_profiles.round(3))

# Plot feature profiles
fig, axes = plt.subplots(2, 5, figsize=(20, 8))
axes = axes.ravel()

feature_names = df_results.drop('cluster', axis=1).columns
colors = plt.cm.tab10(np.linspace(0, 1, len(cluster_profiles)))

for i, feature in enumerate(feature_names):
    ax = axes[i]
    cluster_values = cluster_profiles[feature].values
    bars = ax.bar(range(len(cluster_profiles)), cluster_values, 
                  color=colors, alpha=0.8, edgecolor='black')
    ax.set_title(feature.replace('_', ' ').title())
    ax.set_xticks(range(len(cluster_profiles)))
    ax.set_xticklabels([f'Cluster {i}' for i in range(len(cluster_profiles))])
    ax.tick_params(axis='x', rotation=45)
    
    # Add value labels on bars
    for bar, value in zip(bars, cluster_values):
        height = bar.get_height()
        if abs(height) < 0.1:  # Skip very small values
            continue
        ax.text(bar.get_x() + bar.get_width()/2., height + (0.01 if height >= 0 else -0.03),
                f'{value:.2f}', ha='center', va='bottom' if height >= 0 else 'top', fontsize=8)

plt.suptitle('Average Audio Features by Cluster', fontsize=16, y=1.02)
plt.tight_layout()
plt.savefig('outputs/feature_profiles.png', dpi=300, bbox_inches='tight')
plt.show()

# 3. Cluster interpretation
print("\n" + "="*50)
print("CLUSTER INTERPRETATION")
print("="*50)

# Create cluster descriptions based on feature patterns
cluster_descriptions = {}

for cluster in range(len(cluster_profiles)):
    profile = cluster_profiles.iloc[cluster]
    
    # Analyze key characteristics
    dance_high = profile['danceability'] > 0.6
    dance_low = profile['danceability'] < 0.4
    energy_high = profile['energy'] > 0.6
    energy_low = profile['energy'] < 0.4
    acoustic_high = profile['acousticness'] > 0.7
    valence_high = profile['valence'] > 0.6
    valence_low = profile['valence'] < 0.4
    speech_high = profile['speechiness'] > 0.1
    tempo_fast = profile['tempo'] > 120
    duration_long = profile['duration_ms'] > 200000
    
    # Generate description
    desc_parts = []
    
    if dance_high and energy_high:
        desc_parts.append("🎉 Energetic dance tracks")
    elif dance_high:
        desc_parts.append("💃 Danceable grooves")
    elif energy_high:
        desc_parts.append("⚡ High-energy anthems")
    
    if acoustic_high:
        desc_parts.append("🎸 Acoustic/organic sound")
    elif profile['instrumentalness'] > 0.5:
        desc_parts.append("🎹 Instrumental focus")
    
    if valence_high:
        desc_parts.append("😊 Happy/uplifting")
    elif valence_low:
        desc_parts.append("😔 Melancholic/sad")
    
    if speech_high:
        desc_parts.append("🎤 Vocal-heavy")
    
    if tempo_fast:
        desc_parts.append("🏃 Fast-paced")
    
    if duration_long:
        desc_parts.append("⏰ Extended tracks")
    
    # Default descriptions
    if not desc_parts:
        if energy_low and acoustic_high:
            desc_parts.append("😌 Chill/acoustic vibes")
        elif energy_low:
            desc_parts.append("🌙 Relaxed atmosphere")
        else:
            desc_parts.append("🎵 Standard pop/rock")
    
    cluster_descriptions[cluster] = " | ".join(desc_parts)

# Display descriptions
print("\n🎵 MUSIC CLUSTER PROFILES:")
print("-" * 50)
for cluster, description in cluster_descriptions.items():
    count = df_results[df_results['cluster'] == cluster].shape[0]
    print(f"Cluster {cluster:2d} ({count:4d} songs): {description}")

# Save cluster descriptions
cluster_summary = pd.DataFrame({
    'cluster': list(cluster_descriptions.keys()),
    'description': list(cluster_descriptions.values()),
    'song_count': [df_results[df_results['cluster'] == i].shape[0] for i in cluster_descriptions.keys()]
})
cluster_summary.to_csv('outputs/cluster_summary.csv', index=False)
print(f"\n✅ Cluster summary saved with {len(cluster_summary)} clusters!")

# 4. Top songs per cluster (sample)
print("\n" + "="*50)
print("TOP SONGS PER CLUSTER (Sample)")
print("="*50)

# Merge with original data for track names
df_final_viz = df_viz.merge(df_original[['name_song', 'name_artists']], 
                           left_index=True, right_index=True, how='left')

for cluster in sorted(df_final_viz['cluster'].unique()):
    print(f"\n🎵 Cluster {cluster} - {cluster_descriptions[cluster]}")
    print("-" * 60)
    cluster_songs = df_final_viz[df_final_viz['cluster'] == cluster].head(3)
    for _, song in cluster_songs.iterrows():
        print(f"  🎶 {song['name_song']} by {song['name_artists']}")
    
    # Show average danceability and energy for this cluster
    cluster_data = df_results[df_results['cluster'] == cluster]
    avg_dance = cluster_data['danceability'].mean()
    avg_energy = cluster_data['energy'].mean()
    print(f"     💃 Avg Danceability: {avg_dance:.2f} | ⚡ Avg Energy: {avg_energy:.2f}")

print(f"\n📊 All visualizations saved successfully!")
```
