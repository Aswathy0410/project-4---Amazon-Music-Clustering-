```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

#Step 1: Load Dataset

# Load the dataset (replace with your path if needed)
df = pd.read_csv("single_genre_artists.csv")

# Look at first 5 rows
print(df.head())

# Shape of dataset
print("Shape:", df.shape)

# Column info
print(df.info())

#Step 2: Data Cleaning
# Drop unnecessary columns
cols_to_drop = [
    'id_songs', 'name_song', 'id_artists', 'name_artists', 
    'genres', 'release_date', 'key', 'mode', 'time_signature', 'explicit'
]
df_clean = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

# Check missing values
print("Missing values:\n", df_clean.isnull().sum())

# Drop duplicates if any
df_clean = df_clean.drop_duplicates()

print("Shape after cleaning:", df_clean.shape)
print("Remaining columns:\n", df_clean.columns)

#Step 3: Feature Selection

#We only keep audio features:

features = [
    'danceability', 'energy', 'loudness', 'speechiness', 
    'acousticness', 'instrumentalness', 'liveness', 
    'valence', 'tempo', 'duration_ms', 
    'popularity_songs', 'popularity_artists', 'followers'
]

X = df_clean[features]


#Step 4: Data Normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Step 5: Find Best Number of Clusters (Elbow Method)
sse = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

# Plot Elbow Curve
plt.plot(k_range, sse, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("SSE (Inertia)")
plt.title("Elbow Method")
plt.show()

#Step 6: K-Means Clustering

#(Let’s assume best k=5 after elbow curve)

kmeans = KMeans(n_clusters=5, random_state=42)
df_clean["Cluster"] = kmeans.fit_predict(X_scaled)

#Step 7: Cluster Evaluation
score = silhouette_score(X_scaled, df_clean["Cluster"])
print("Silhouette Score:", score)

#Step 8: Visualization with PCA (2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=df_clean["Cluster"], cmap="rainbow", alpha=0.7)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Song Clusters (PCA Visualization)")
plt.show()

#Step 9: Cluster Analysis
cluster_profile = df_clean.groupby("Cluster")[features].mean()
print(cluster_profile)

# Heatmap for better visualization
plt.figure(figsize=(10,6))
sns.heatmap(cluster_profile, annot=True, cmap="coolwarm")
plt.title("Average Feature Values per Cluster")
plt.show()

#Step 11: Save Final Results
df_clean.to_csv("clustered_songs.csv", index=False)
print("Final dataset with clusters saved as clustered_songs.csv")



```
